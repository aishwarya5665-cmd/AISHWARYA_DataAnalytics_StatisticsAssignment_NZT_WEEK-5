**TASK 1-STATISTICAL ANALYSIS**

**1. Descriptive Statistics**
Perform statistical analysis on your dataset:
• Calculate: Mean, Median, Mode (any one numerical column)
• Find: Range & Standard Deviation
• Create a Frequency Distribution Table
• Plot Histogram of any numerical data
➤ Purpose – Understand central tendency & spread of data.

**Importing Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

**Loading Dataset**

```python
df = pd.read_excel('/StudentScores.xlsx')
df
```

```python
col = "Math_Score"
```

**Computing Statistics**

```python
mean_val = df[col].mean()
median_val = df[col].median()
mode_val = df[col].mode()[0]
range_val = df[col].max() - df[col].min()
std_dev = df[col].std()
```

```python
stats = {
    "Mean": mean_val,
    "Median": median_val,
    "Mode": mode_val,
    "Range": range_val,
    "Standard Deviation": std_dev
}
```

```python
#Frequency distribution
freq_dist = df[col].value_counts().sort_index()
```

```python
#Plotting histogram
plt.figure(figsize=(8,5))
plt.hist(df[col])
plt.title("Histogram of Math Scores")
plt.xlabel("Math Score")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()

stats, freq_dist.head()
```

**Final Insights**

**Overall Math Performance Is Above Average**

The mean Math score is 70.6, which indicates that most students perform better than the passing/average mark (typically 50 or 60).

The median (72) confirms that half the students score above 72 — a good indicator of consistent performance.*

**Score Distribution Is Fairly Balanced**

Mean (70.6), Median (72), and Mode (72) are close →
✔ No extreme skewness
✔ Scores cluster around 70–75

This suggests a healthy distribution without extreme outliers pulling the mean up or down.

**Students Scores Vary Widely**

Standard deviation is 16.5 → quite high.
It states that:

Some students score extremely high- 90+

Some score low -below 50

The range of 59 points  shows a large performance gap.

 Implication:
Teaching strategies may need to be differentiated — some students may require extra support, while high performers may need enrichment tasks.

**Majority of Students Score Between 60 and 90**
In the histogram:
Most bars fall in the middle range.
Very few are in the low 40–50 range, and a handful are in the 90+ range.

Implication:
Class performance is overall strong, but improvement for weak performers is possible.

**No Indication of Data Issues or Anomalies**
Distribution is smooth, without spikes or gaps.
No clustering at boundaries like exactly 0 or 100.
This suggests the dataset is clean and realistic.

**FINAL CONCLUSION**
The Math score analysis reveals that :
Students perform at a generally good level, with most scoring around 70–75, but the presence of high variability suggests the need for targeted interventions for low scorers and opportunities for advanced learners.

```python
#Scatter plot
plt.figure(figsize=(7,5))
plt.scatter(x, y)
plt.xlabel("Marketing Spend")
plt.ylabel("Units Sold")
plt.title("Scatter Plot: Marketing Spend vs Units Sold")
plt.tight_layout()
plt.show()
```

```python
corr
```

**Interpret Result**

**Correlation Result**

**Interpretation**

Weak correlation (because |0.12| < 0.3)

Positive correlation (because r > 0)

**Interpretation of the Scatter Plot**
**Weak Relationship**
The points are scattered loosely, not forming a line.
→ Marketing spend does not strongly determine sales quantity.

**Positive Direction**
General tendency:
Higher spend - slightly higher units sold
But not consistent or strong.

**Possible Explanation**
Because the dataset is randomly generated, and marketing spend values were not linked to sales during creation, a low correlation is expected

**FINAL INSIGHTS**
There is a weak positive correlation (r = 0.1286) between marketing spend and units sold.
While higher marketing spending shows a small tendency to increase sales, the relationship is not strong enough to predict sales accurately.

**TASK 4-REGRESSION ANALYSIS**

**Importing Libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np
```

**Loading dataset**

```python
df = pd.read_excel('/content/Health Data.xlsx')
df
```

```python
X = df[["Age"]]
Y = df["Blood_Pressure"]
```

```python
model = LinearRegression()
model.fit(X, Y)
```

```python
m = model.coef_[0]
c = model.intercept_
```

```python
new_age = 50
predicted_bp = model.predict([[new_age]])[0]
```

```python
plt.figure(figsize=(7,5))
plt.scatter(X, Y)
plt.plot(X, model.predict(X))
plt.xlabel("Age")
plt.ylabel("Blood Pressure")
plt.title("Regression Line: Age vs Blood Pressure")
plt.tight_layout()
plt.show()
```

```python
m, c, predicted_bp
```

**INTERPRETATION**
**Type of Relationship:**

Weak Positive Linear Relationship

As Age increases, Blood Pressure tends to increase slightly.

Slope is small (0.1014) → Increase in age contributes mildly to blood pressure.

**Strength of Relationship**

Slight upward trend, not steep

Blood pressure is influenced by more factors than age alone
(e.g., lifestyle, medical conditions, fitness).

**FINAL INSIGHTS**
Component	Result
Model	-Simple Linear Regression
X (Independent Variable)-	Age
Y (Dependent Variable)-	Blood Pressure
Equation	BP = 0.1014 × Age + 138.8149
Prediction (Age = 50)-	143.88 mmHg
Relationship-Weak Positive
Plot	Successfully created

**TASK 5-HYPOTHESIS TESTING-DECISION MAKING(ONE-WAY ANOVA)**

**Importing Libraries**

```python
import pandas as pd
from scipy.stats import f_oneway
```

**Loading dataset**

```python
df = pd.read_excel('/Employee Salary.xlsx')
df
```

```python
#Salary groups by department
groups = [df[df["Department"]==dept]["Salary"] for dept in df["Department"].unique()]
```

```python
#Performing  ONE WAY ANOVA TEST
anova_result = f_oneway(*groups)
```

```python
anova_result
```

**FINAL INSIGHTS**
Since
p_value=0.6807>0.05
Failed to reject the Null Hypothesis.

There is no statistically significant difference in the average salary across different departments.
